```python
xx_response = df[df.Q21 == 'XXXXXX'].iloc[0]  # name of expert blinded for code release
xx_response
```

```python
df.iloc[0]
```

```python
resp_questions = ['Q1', 'Q19', 'Q11', 
'Q14', 'Q31', 'Q34', 'Q37', 'Q40', 'Q44', 'Q46', 'Q48', 'Q51', 'Q54', 'Q57', 'Q60', 'Q63', 
'Q66', 'Q25']
phase_questions = ['Q83', 'Q75', 'Q76', 'Q77', 'Q78', 'Q79',]
```
